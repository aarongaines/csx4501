```python
import sklearn.datasets as ds
```

```python
diabetes = ds.load_diabetes()
```

```python
print(diabetes.DESCR)
```

```python
x = diabetes.data
y = diabetes.target
```

```python
print(x.shape)
print(y.shape)
```

```python
col_num = x.shape[1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c_list = []
for i in range(col_num):
    corr = np.corrcoef(x[:, i], y)[0, 1]
    c_list.append(corr)
    plt.figure(figsize=(6, 6))
    plt.scatter(x[:, i], y)
    plt.title("Correlation of column {}: {}".format(i+1, corr))
```

```python
c_list = np.array(c_list)
```

```python
np.amax(c_list)
```

```python
print(np.where(c_list == np.amax(c_list))[0])
```

```python
x_1 = x[:,2]
```

```python
plt.scatter(x_1, y)
```

```python
import sklearn.linear_model as lm
import sklearn.model_selection as ms
import sklearn.metrics as metrics
```

```python
model = lm.LinearRegression()


x_train, x_test, y_train, y_test =ms.train_test_split(
    x_1, y, test_size=0.2)


model.fit(x_train.reshape(-1,1),
          y_train)


plt.scatter(x_train,y_train,color='black')
plt.scatter(x_test,y_test,color='blue')


y_pred = model.predict(x_test.reshape(-1,1))

plt.plot(x_test,y_pred,'red')


plt.show()


print('MSE_linreg = %.2f' % metrics.mean_squared_error(y_test, y_pred))
print('R2_linreg = %.2f' % metrics.r2_score(y_test, y_pred))
```
